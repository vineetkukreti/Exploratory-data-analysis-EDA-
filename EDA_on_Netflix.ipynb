```python
import pandas as pd

def make_proper_date(data):
    # Remove leading and trailing whitespace from the input data
    data = data.strip()
    
    date_format = "%B %d, %Y"
    try:
        pandas_datetime = pd.to_datetime(data, format=date_format)
        return pandas_datetime
    except ValueError:
        return None
df['date_added'] = df['date_added'].apply( make_proper_date)
```
