```python
df.dropna(inplace=True)
```

```python
df['DJ_Date']  = df['Date_of_Journey'].str.split('/').str[0]
df['DJ_Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['DJ_Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['DJ_Date'] = df['Date'].astype(int)
df['DJ_Month'] = df['Month'].astype(int)
df['DJ_Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace = True)
```

```python
df.head(2)
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(' ').str[0].str.split(":").str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(' ').str[0].str.split(":").str[1]
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python

df['Dep_Time_Hour'] = df['Dep_Time'].str.split(' ').str[0].str.split(":").str[0]
df['Dep_Time_Minute'] = df['Dep_Time'].str.split(' ').str[0].str.split(":").str[1]
df['Dep_Time_Hour'] = df['Dep_Time_Hour'].astype(int)
df['Dep_Time_Minute'] = df['Dep_Time_Minute'].astype(int)
```
