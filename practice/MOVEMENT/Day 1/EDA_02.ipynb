```python
fig,axis  = plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='average',bins=30,kde=True,color='g')

plt.subplot(122)
sns.histplot(data=df,x='average',bins = 30,kde=True,hue='gender')
```

```python
#Insight from the graph -> it tells that female student is more studious than male student which clearly can be seem from the graph
#Also from subplot 121 and subplot 122 ,both are seems to be a bell shaped graph [Normallay distributed graph/ Gaussian Graph]
```

```python
plt.subplots(1,2,figsize=(25,7))
plt.subplot(141)
sns.histplot(data=df,x='average',kde=True,hue='lunch')

plt.subplot(142)
sns.histplot(data=df[df.gender=='female'],x='average',kde=True,hue='lunch')

plt.subplot(143)
sns.histplot(data=df[df.gender=='female'],x='average',kde=True,hue='lunch')
```

Insights
* standard lunch helps students perform well in exams
* standard lunch helps perform well in exams be it a male or female

```python
plt.subplots(1,2,figsize=(40,7))
plt.subplot(141)
sns.histplot(data=df,x='average',kde=True,hue='parental_level_of_education').set_title("How parental_level_of_education is effecting the average score of student")

plt.subplot(142)
sns.histplot(data=df[df.gender=='female'],x='average',kde=True,hue='parental_level_of_education').set_title("How parental_level_of_education is effecting the average score of female student")


plt.subplot(143)
sns.histplot(data=df[df.gender=='male'],x='average',kde=True,hue='parental_level_of_education').set_title("How parental_level_of_education is effecting the average score of male student")
```

Insight for the graph
* In 1st plot, general parent's educaion don't help student perform well in exam
* In 2nd plot ,we can see there is no effect of parent's education on female students.
* 3rd plot shows that's parent whose education is of associative's degree or master's degree their male child tend to perform well in exam


```python
plt.subplots(1,2,figsize=(40,7))
plt.subplot(141)
sns.histplot(data=df,x='average',kde=True,hue='race_ethnicity').set_title("How race_ethnicity is effecting the average score of student")

plt.subplot(142)
sns.histplot(data=df[df.gender=='female'],x='average',kde=True,hue='race_ethnicity').set_title("How race_ethnicity is effecting the average score of female student")


plt.subplot(143)
sns.histplot(data=df[df.gender=='male'],x='average',kde=True,hue='race_ethnicity').set_title("How race_ethnicity is effecting the average score of male student")
```

Insight 
* students of group A and group B tends to perform poorly in exam.
